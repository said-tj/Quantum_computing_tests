```python
# Ejemplo auto-contenible para un notebook de Jupyter Lab
#
# Muestra:
#   • Una breve descripción textual de cada compuerta cuántica
#   • El circuito generado
#   • El vector de estado resultante después de aplicar la compuerta
#
# Requiere Qiskit >= 0.43.  Si no está instalado, descomenta la primera
# línea para instalarlo desde el propio notebook.  Cada bloque se puede
# ejecutar de manera independiente y sirve como plantilla para añadir
# más compuertas o ejemplos.

# !pip install qiskit --quiet

from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer
from qiskit.quantum_info import Statevector
import numpy as np

backend = Aer.get_backend("statevector_simulator")


def run_and_show(qc, description=""):
    """Ejecuta el circuito en el simulador de vectores de estado y muestra resultados."""
    if description:
        print(description)
    print(qc.draw("text"))
    state = Statevector.from_instruction(qc)
    print("Vector de estado:", state.data, "\n")


# --------------------------------------------------------------------
# 1-Qubit Gates
# --------------------------------------------------------------------
print("-" * 60)
print("COMPuertas de un qubit\n")

# --- X (NOT) ---
qc_x = QuantumCircuit(1, name="X")
qc_x.x(0)
run_and_show(
    qc_x,
    "X — NOT cuántico.\nInvierte |0⟩↔|1⟩.   X|0⟩ = |1⟩,   X|1⟩ = |0⟩",
)

# --- H (Hadamard) ---
qc_h = QuantumCircuit(1, name="H")
qc_h.h(0)
run_and_show(
    qc_h,
    "H — Hadamard.\nCrea superposición: H|0⟩ = (|0⟩+|1⟩)/√2",
)

# --- Z ---
qc_z = QuantumCircuit(1, name="Z")
qc_z.z(0)
run_and_show(
    qc_z,
    "Z — Pauli-Z.\nAñade fase −1 al estado |1⟩: Z|1⟩ = −|1⟩;  deja |0⟩ intacto.",
)

# --- Y ---
qc_y = QuantumCircuit(1, name="Y")
qc_y.y(0)
run_and_show(
    qc_y,
    "Y — Pauli-Y.\nCombina X y Z con fase imaginaria.  Gira el qubit en la esfera de Bloch ±π alrededor de Y.",
)

# --- S (phase π/2) ---
qc_s = QuantumCircuit(1, name="S")
qc_s.s(0)
run_and_show(
    qc_s,
    "S — Phase gate (π/2).\nAñade fase i al estado |1⟩.",
)

# --- T (phase π/4) ---
qc_t = QuantumCircuit(1, name="T")
qc_t.t(0)
run_and_show(
    qc_t,
    "T — Phase gate (π/4).\nAñade fase e^{iπ/4} al estado |1⟩.",
)

# --------------------------------------------------------------------
# 2-Qubit Gate — CNOT
# --------------------------------------------------------------------
print("-" * 60)
print("Compuerta de dos qubits\n")

qc_cnot = QuantumCircuit(2, name="CNOT")
qc_cnot.h(0)          # Prepara superposición en el qubit de control
qc_cnot.cx(0, 1)      # CNOT control=0 -> target=1
run_and_show(
    qc_cnot,
    "CNOT — Control-NOT.\nSi el control está en |1⟩ aplica X al qubit objetivo.\n" +
    "Ejemplo con H en el control: genera estado Bell (|00⟩+|11⟩)/√2.",
)


```

```python
# !pip install "qiskit>=0.46" qiskit-aer --quiet

%matplotlib inline

from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_state_city
from IPython.display import display, Markdown
import numpy as np

# ------------------------------------------------------------------
#  backend simulador
# ------------------------------------------------------------------
backend = Aer.get_backend("aer_simulator")   # incluye statevector_simulator

# ------------------------------------------------------------------
#  helper para mostrar todo ordenado
# ------------------------------------------------------------------
def run_and_show(qc: QuantumCircuit, title: str):
    """Imprime descripción, diagrama (matplotlib) y estado final."""
    display(Markdown(f"## {title}"))
    display(qc.draw("mpl"))
    state = Statevector.from_instruction(qc)
    display(Markdown(f"**Vector de estado:** `{np.round(state.data,4)}`"))
    #  visual extra
    if qc.num_qubits == 1:
        display(plot_bloch_multivector(state))
    else:
        display(plot_state_city(state, title="Amplitudes"))

# ------------------------------------------------------------------
# 1-Qubit gates
# ------------------------------------------------------------------
qc_x = QuantumCircuit(1, name="X");  qc_x.x(0)
run_and_show(qc_x, "X — NOT cuántico  \nInvierte |0⟩ ↔ |1⟩")

qc_h = QuantumCircuit(1, name="H");  qc_h.h(0)
run_and_show(qc_h, "H — Hadamard  \nCrea superposición (|0⟩+|1⟩)/√2")

qc_z = QuantumCircuit(1, name="Z");  qc_z.z(0)
run_and_show(qc_z, "Z — Pauli-Z  \nAñade fase −1 a |1⟩")

qc_y = QuantumCircuit(1, name="Y");  qc_y.y(0)
run_and_show(qc_y, "Y — Pauli-Y  \nGiro π alrededor del eje Y")

qc_s = QuantumCircuit(1, name="S");  qc_s.s(0)
run_and_show(qc_s, "S — Phase π/2  \nAñade fase +i al estado |1⟩")

qc_t = QuantumCircuit(1, name="T");  qc_t.t(0)
run_and_show(qc_t, "T — Phase π/4  \nAñade fase e^{iπ/4} a |1⟩")

# ------------------------------------------------------------------
# 2-Qubit: CNOT + Bell
# ------------------------------------------------------------------
qc_cnot = QuantumCircuit(2, name="CNOT")
qc_cnot.h(0)        # superpone qubit 0
qc_cnot.cx(0, 1)    # CNOT control 0 → target 1
run_and_show(
    qc_cnot,
    "CNOT — Control-NOT  \nCon H previo genera el estado Bell (|00⟩+|11⟩)/√2"
)
```
